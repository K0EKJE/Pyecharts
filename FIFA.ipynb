```python
def barplot(x_data, y_data, pos):
    bar = (Bar(init_opts=opts.InitOpts(width='1000px'
                                        , height='600px'
                                        , theme='vintage'
                                        ))
           .add_xaxis(x_data)
           .add_yaxis('',y_data)
           .set_series_opts(label_opts=opts.LabelOpts(
                                                      position="insideRight"
                                                      #, font_size=12
                                                      ,formatter='{c}'
                                                      #,color = 'white'
                                                      )
                          )
           # 标签位置 'top'，'left'，'right'，'bottom'，'inside'，'insideLeft'，'insideRight'
                                               # 'insideTop'，'insideBottom'， 'insideTopLeft'，'insideBottomLeft'
                                               # 'insideTopRight'，'insideBottomRight'

           .set_global_opts(
                            xaxis_opts=opts.AxisOpts(is_show=True
                                                     ,name_gap = 40
                                                     ,name='Number of Players'
                                                     ,axistick_opts=opts.AxisTickOpts(is_inside=False)
                                                     ,name_location = 'center'
                                                     ,name_textstyle_opts = opts.TextStyleOpts(font_weight='bold'
                                                                                               ,font_size=14
                                                                                               ,font_family='Times'
                                                                                               ,color='#292421')
                                                     )
                           ,yaxis_opts=opts.AxisOpts(is_show=True
                                                     ,is_inverse = True
                                                     ,name='Country'
                                                     ,name_location = 'start'
                                                     ,name_textstyle_opts = opts.TextStyleOpts(font_weight='bold'
                                                                                               ,font_size=14
                                                                                               ,font_family='Times'
                                                                                               ,color='#292421')
                                                     ,axislabel_opts = opts.TextStyleOpts(font_weight='bold'
                                                                                               ,font_size=12
                                                                                               ,font_family='Times'
                                                                                               ,color='#B22222')
                                                     ,axisline_opts=opts.AxisLineOpts(is_show=True,
                                                     linestyle_opts=opts.LineStyleOpts(width=2, color='black'))
                                                    )


                   
                            ,title_opts=opts.TitleOpts(
                                     title=f'Top 10 Countries With Most: {pos}',
                                     title_textstyle_opts=opts.TextStyleOpts(#font_style="italic",
                                                                             color='#292421',
                                                                             font_weight='bold',
                                                                             font_family='Times',
                                                                             font_size=25),
                                                                             pos_left='20%',
                                                                             pos_top = '20',
                                                       )
                          ,visualmap_opts=opts.VisualMapOpts(is_show=False
                                                             #,max_ = 3000
                                                             ,range_color=['#B0171F']
                                                             )
                           )
          )
    bar.reversal_axis()
    return bar
```

```python
def radar_player(names,skills_opt=[],label = True,mode = 'multiple'):
    '''
    Input: 
    ------
    Names : a list containing the specific searching targets
    Skills_opt : attributes to search; passed by a list or a number from 0-6 corresponding to pre-decided
                 attribute sets for Attacking,Skill,Movement,Power,Mentality,Defending,Goalkeeping
    label : default = True, takes value 'off' and 'show'
    mode  : 'multiple','single'; controls whether display the radar of multiple players
    ------
    Output: a radar graph
    '''
    #colors = ['#0000FF','#33A1C9','#708069','#082E54','#2E8B57','#9933FA','#B0171F','#FF9912','#1E90FF','black']
    colors = ['#00C78C','#1E90FF','#9933FA','#B0171F','#5E2612','#DA70D6','#0B1746','black','#9933FA']
     
    Attacking = ['Crossing','Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys']
    Skill = ['Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control']
    Movement = ['Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance']
    Power = ['Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots']
    Mentality = ['Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure']
    Defending = ['Standing Tackle', 'Sliding Tackle','Defensive Awareness', 'Interceptions','Interceptions','Strength'] 
    Goalkeeping = ['GK Diving', 'GK Handling','GK Kicking', 'GK Positioning', 'GK Reflexes']
    
    if type(skills_opt)!=list:
        skills_opt = [Attacking,Skill,Movement,Power,Mentality,Defending,Goalkeeping][skills_opt]
    
    if label == 'off': # turn off label if explicitly specified, otherwise turn of
        label = False

    c_schema = []
    for i in range(len(skills_opt)):
        c_schema.append(opts.RadarIndicatorItem(name=f"{skills_opt[i]}", max_=100, color = colors[i%9]))

    attr = np.array(df[skills_opt])


    radar = (Radar(init_opts=opts.InitOpts(width="1200px", height="700px", theme = 'vintage'))
             .add_schema(schema = c_schema
                         ,shape = 'circle'
                         ,splitarea_opt=opts.SplitAreaOpts(
                                                        is_show= True
                                                        ,areastyle_opts=opts.AreaStyleOpts(opacity=1)
                                                        )
                         ,textstyle_opts=opts.TextStyleOpts(color="#000000"
                                                         ,font_size=20),
                        )
            )
    
    border = (2,'red')
    if mode == 'single':border = (5,'red')
        
    for i in range(len(names)):

        value = attr[df.Name == names[i]].tolist()
        radar.add(
                    series_name=f"{names[i]}",
                    data=value,
                    areastyle_opts=opts.AreaStyleOpts(color=colors[i%9],opacity=0.4), # 区域面积，透明度
                    linestyle_opts=opts.LineStyleOpts(width=border[0],color=border[1])
                    )
             
    radar.set_series_opts(label_opts=opts.LabelOpts(position='insideBottom',is_show=label
                                                    #,color = 'black'
                                                    ,font_family = 'Arial'
                                                    ,font_size = 14))
    radar.set_global_opts(
                    title_opts=opts.TitleOpts(title = "Player Attributes Radar"
                                             #,subtitle = f'Conditions: {}'
                                             ,pos_top = '3%'
                                             ,pos_left = '2%'
                                             ,title_textstyle_opts=opts.TextStyleOpts(color='#292421',
                                                                                      font_weight='bold',
                                                                                      font_family='Times',
                                                                                      font_size=25)
                                              ,subtitle_textstyle_opts=opts.TextStyleOpts(color='#808A87',
                                                                                      font_weight='bold',
                                                                                      font_family='Times',
                                                                                      font_size=18), 
                                             ),        

                    legend_opts=opts.LegendOpts(pos_top = '30%'
                                                #,pos_right = '10'
                                                ,pos_left = '10'
                                                ,type_ = 'scroll'
                                                ,align = 'auto'
                                                ,orient = 'vertical'
                                                ,legend_icon = 'none'
                                                ,inactive_color = 'lightgrey'
                                                ,textstyle_opts = opts.TextStyleOpts(color='blue')                  
                                                ,selected_mode = mode
                                                ,item_height = 10
                                                ),  

                    )
    
    grid = (Grid(init_opts=opts.InitOpts(width="1200px", height="600px", theme = 'vintage'))
            .add(radar,grid_opts=opts.GridOpts(pos_left="1%"))
            )
    return grid
```

```python
def combo_radar(names, skills_opt = [], label = 'show', mode = 'single'):
    '''
    Input: 
    ------
    Names : a list containing the specific searching targets
    Skills_opt : attributes to search; passed by a list or a number from 0-6 corresponding to pre-decided
                 attribute sets for Attacking,Skill,Movement,Power,Mentality,Defending,Goalkeeping
    label : default = True, takes value 'off' and 'show'
    mode  : 'multiple','single'; controls whether display the radar of multiple players
    ------
    Output: a set of radar graphs
    '''
    tab = Tab()
    if len(skills_opt)>0:
        radar = radar_player(names = names
                         ,skills_opt = skills_opt
                         ,label = label
                         ,mode = mode)
        tab.add(radar, 'Chosen features')
    
    tabs = ['Attacking', 'Skill', 'Movement', 'Power', 'Mentality', 'Defending', 'Goalkeeping']
    for i, tab_name in enumerate(tabs):
        radar = radar_player(names = names
                             ,skills_opt = i
                             ,label = label
                             ,mode = mode)
        tab.add(radar, tab_name)
        
    return tab
```

```python
# Transform the two columns to get useful info
df['Weak Foot'] = df['Weak Foot'].apply(lambda x:x[:3])
df['Weight'] = df['Weight'].apply(lambda x: int(x[:3]))
df['Height_cm'] = df['Height'].apply(lambda x: ((int(x.split("'")[0]))*12+int(x.split("'")[1]))*2.54)
```
